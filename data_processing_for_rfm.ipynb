### Masking on Ha Oey Data

```python
import pandas as pd 
import numpy as np 
```

```python
haoey_data = pd.read_csv('haoey_data_processed.csv')
haoey_data
```

```python
# Get unique values of 'Provider' column
unique_providers = haoey_data['Patient ID'].unique()

# Create a new list called provider_masked
provider_masked = [f'Patient_{i}' for i in range(1, len(unique_providers)+1)]

# Create a dictionary to map original provider names to masked names
provider_mapping = dict(zip(unique_providers, provider_masked))

# Mask the 'Provider' column with the masked names
haoey_data['Patient ID'] = haoey_data['Patient ID'].map(provider_mapping)

haoey_data
```

```python
# Get unique values of 'Provider' column
unique_providers = haoey_data['Provider'].unique()

# Create a new list called provider_masked
provider_masked = [f'Provider_{i}' for i in range(1, len(unique_providers)+1)]

# Create a dictionary to map original provider names to masked names
provider_mapping = dict(zip(unique_providers, provider_masked))

# Mask the 'Provider' column with the masked names
haoey_data['Provider'] = haoey_data['Provider'].map(provider_mapping)

haoey_data
```

```python
haoey_data.to_csv('haoey_data_processed.csv', index = False)
```

```python
import numpy as np
import pandas as pd

# Convert 'Date' column to datetime if it's not already
haoey_data['Date'] = pd.to_datetime(haoey_data['Date'])

# Initialize an empty list to store new rows
new_rows = []

# Iterate over each row
for _, row in haoey_data.iterrows():
    new_rows.append(row)  # Always keep the original row
    
    # Check if we should duplicate this row
    if np.random.rand() > 0.7:
        duplicated_row = row.copy()
        duplicated_row['Date'] += pd.Timedelta(days=1)  # Increment date by 1 day
        new_rows.append(duplicated_row)

# Create a new DataFrame with the original and duplicated rows
haoey_data = pd.DataFrame(new_rows)

# Reset index
haoey_data.reset_index(drop=True, inplace=True)

# Display the new dataset
haoey_data
```

```python
haoey_data.to_csv('haoey_data_processed.csv', index = False)
```
